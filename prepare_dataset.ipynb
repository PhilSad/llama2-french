```python
from datasets import load_dataset
```

```python
data = load_dataset('Enno-Ai/fr-instructs', split='train')
```

```python
def concatenate_columns():
    row["concatenated_colurowmn"] = f"###User: {row['instruction']} {row['input']} ###Assistant: {row['output']}"
    return row

data_mapped = data.map(concatenate_columns)
```

```python
data_mapped = data_mapped.remove_columns(['instruction', 'input', 'output', 'source'])
```

```python
data_mapped = data_mapped.rename_column('concatenated_column', "text")
```

```python
data_mapped.push_to_hub("PhilSad/Instruct-fr-merged")
```
